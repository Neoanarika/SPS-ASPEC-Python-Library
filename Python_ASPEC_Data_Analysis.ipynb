# Atomic Spectra Library

```python
import numpy as np
import matplotlib.pyplot as plt
from aspec import SpectralData, AngleData, ProcessedSpectralData
```

```python
plt.style.use('ggplot')
```

```python
spectral = SpectralData('./Lab 3 Results/180919 X 2.txt')
angle = AngleData("./Lab 3 Results/180919 X 2 ang.txt")
```

```python
X, freqs, f_s = spectral.fft()
```

```python
fig, ax = plt.subplots()

ax.stem(freqs, np.abs(X))
ax.set_xlabel('Frequency in Hertz [Hz]')
ax.set_ylabel('Frequency Domain (Spectrum) Magnitude')
ax.set_xlim(0, f_s / 2)
```

```python
data = ProcessedSpectralData(spectral, angle, weight=0.2, cutoff_freq=1.8)
```

```python
plt.plot(data.angle, data.light_intensity)
plt.title('Kellisa Hydrogen')
plt.ylabel('light intensity')
plt.xlabel('revolution')
```

```python
plt.plot(data.theta, data.light_intensity)
plt.title('Kellisa Hydrogen')
plt.ylabel('light intensity')
plt.xlabel('angle (degree)')
```
